# This is highly unbalanced

```python
# Seprating for the data analysis

ligit = credit_card_data[credit_card_data.Class == 0]
fruad = credit_card_data[credit_card_data.Class == 1]
```

# Under sampling 
Build a sample dataset containing similar distribution of normal and fruadulent transactions

number of fruadulent transaction is 492

```python
ligit_sample = ligit.sample(n = 492)
```

Concenate two DataFrames

```python
new_dataset = pd.concat([ligit_sample, fruad], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# spliting the data into Features & Targets

```python
X = new_dataset.drop(columns ='Class', axis = 1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

# Spliting the data into Training Data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y , random_state = 2)
```

```python
X.shape, X_train.shape, X_test.shape
```

# Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the logistic Regression Model with Training Data 
model.fit(X_train, Y_train)
```

# Model Evaluation 
Accuracy Score

```python
# Accuracy with training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
'Accuracy on Training data :',training_data_accuracy
```

```python
# Accuracy with testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
('Accuracy on Testing Data :',test_data_accuracy)
```
